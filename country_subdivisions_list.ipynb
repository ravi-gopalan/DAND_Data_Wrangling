```
import pandas as pd
```

```
list(pycountry.countries)
```

```
df_countries = pd.DataFrame([{'country_alpha_2': country.alpha_2,'country_alpha_3': country.alpha_3, 'country_name': country.name.lower(), 'country_numeric':country.numeric}  for country in list(pycountry.countries)])
df_countries.info()
```

```
df_countries.head()
```

```
list(pycountry.subdivisions)[:2]
```

```
country_alpha_2_list = list(df_countries['country_alpha_2'].unique())
```

```
country_alpha_2_list[0:10]
```

```
df = pd.DataFrame()
for country_code in country_alpha_2_list:
  df = pd.concat([df,pd.DataFrame([{'country_alpha_2': country_code,'sd_code': sd.code, 'sd_name': sd.name.lower(), 'sd_type':sd.type, 'sd_parent_code':sd.parent_code}  for sd in list(pycountry.subdivisions) if sd.country_code == country_code])])
df.info()
```

```
df.tail()
```

```
df[df.sd_parent_code.notna()]
```

```
df_countries.to_csv('country_detail.csv',index=False)
df.to_csv('country_subdivisions_detail.csv',index=False)
```
